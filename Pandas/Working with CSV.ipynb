```python
# lets create a best list of movies and save it in another csv

df2 = df.sort_values(['IMDb','BoxOffice(INR Crore)'], ascending = False )
```

```python
# we have to reset the index , not to sort so we'll use reset
df2.reset_index(drop=True, inplace = True )
```

```python
# lets just save top 5 movies 
Top_5_df = df2.head()
```

```python
# now to save it in new csv
Top_5_df.to_csv('Top_5_Film.csv')
```

```python
# let's save actors name in another excel sheet
Actor_df = Top_5_df['Actor']
```

```python
with pd.ExcelWriter("report.xlsx") as writer:
    Top_5_df.to_excel(writer, sheet_name="TOP MOVIES", index=False)
    Actor_df.to_excel(writer, sheet_name="BEST ACTORS", index=False)
```
