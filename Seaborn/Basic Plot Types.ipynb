# Plotting graphs


```python
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (6, 4)
```

```python
# 1. line graph
import matplotlib.pyplot as plt
sns.set_theme(style='ticks')
plt.rcParams["figure.figsize"] = (6, 4)
plt.figure(figsize=(6, 4))
sns.lineplot(x = 'sepal_length' ,y = 'sepal_width', data = flower )
plt.show()
```

```python
# 2.Scatter Plot
sns.scatterplot(x='petal_width', y='petal_length', data=flower)
plt.show()
```

```python
# 3.Bar plot
sns.barplot(x='sepal_width',y='sepal_length', data=flower)
```
