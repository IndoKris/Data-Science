# Plotting graphs


```python
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (6, 4)
```

```python
# 1. line graph
import matplotlib.pyplot as plt
sns.set_theme(style='ticks')
plt.rcParams["figure.figsize"] = (6, 4)
plt.figure(figsize=(6, 4))
sns.lineplot(x = 'sepal_length' ,y = 'sepal_width', data = flower )
plt.show()
```

```python
# 2.Scatter Plot
sns.scatterplot(x='petal_width', y='petal_length', data=flower)
plt.show()
```

```python
# 3.Bar plot
sns.barplot(x='sepal_width',y='sepal_length', data=flower)
plt.show()
```

```python
df = sns.load_dataset('tips')
```

```python
# 4.Boxplots
sns.boxplot(x='tip',y='day',hue='smoker' data=df)
```

```python
# 5.Heatmap (Correlation Matrix)
flights = sns.load_dataset('flights')
pivot_table = flights.pivot(index="month",columns= "year",values= "passengers")
sns.heatmap(pivot_table, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Heatmap of Passengers')
plt.show()
```

```python
# annot means annotation 
# fmt means format 
# cmap means colormap
```
