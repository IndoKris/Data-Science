# Here we'll learn more about charts 

```python
# Using Format Strings
import matplotlib.pyplot as plt

years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
plt.plot(years, kohli,'ro--', label="Virat Kohli")
plt.plot(years, sehwag, 'g^:', label="Virender Sehwag")
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend()
plt.show()
```

```python
plt.plot(years,kohli,color='cyan',linestyle = ':',label = 'Virat Kohli')
plt.plot(years,sehwag , color = 'yellow', linestyle='-.',label = 'Virender Sehwag')
plt.legend()
plt.show()
```

```python
a = plt.plot(years,kohli,color='cyan',linestyle = ':',label = 'Virat Kohli')
b = plt.plot(years,sehwag , color = 'yellow', linestyle='-.',label = 'Virender Sehwag')

```

```python
a = plt.plot(years,kohli,color='green',linestyle = ':',label = 'Virat Kohli')
b = plt.plot(years,sehwag , color = 'purple', linestyle='-.',label = 'Virender Sehwag')
plt.grid(True)
plt.tight_layout()
plt.legend()
plt.show()
plt.style.use('seaborn-v0_8-darkgrid')
```

## Assignment 
- Create a plot comparing Kohli, Rohit Sharma, and Sehwag across 10 years of hypothetical runs.
- Labels
- Legends
- Colors
- Line styles
- One custom style

```python
import matplotlib.pyplot as plt 

years = [2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018]

kohli_runs = [325, 995, 1381, 1026, 1268, 1054, 1380, 1215, 1460, 1202]

rohit_runs = [362, 405, 611, 563, 1196, 578, 815, 930, 1293, 1030]

sehwag_runs = [1236, 1462, 880, 1115, 1041, 754, 651, 384, 217, 96]

plt.plot(years,kohli_runs ,color='red' ,linestyle =':',linewidth= 2 ,label = 'Virat Kohli')
plt.plot(years,rohit_runs ,color='green',linestyle = '--',linewidth= 1 , label = 'Rohit Sharma')
plt.plot(years,sehwag_runs,color='blue',linestyle='-' ,linewidth= 3 , label = 'Virender Sehwag') 
plt.title("Performance Comparison")
plt.xlabel('Years')
plt.ylabel('Players')
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.style.use('grayscale')
plt.show()
```
