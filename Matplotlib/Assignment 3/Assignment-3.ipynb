```python
#imports
import matplotlib.pyplot as plt 

#data with others expanded 
data = [72, 27, 0.4, 0.3, 0.2, 0.1]
labels = ['Android', 'iOS', 'KaiOS', 'Windows Phone', 'Samsung', 'Other']
colors = ['#304852','#36595F','#5F818A','#405559','#C7EEFF','#B9DBE6']
#pie chart
plt.pie(
    x = data,
    labels = labels,
    colors = colors,
    shadow = True,
    autopct = '%.1f%%',
    frame = False,
    wedgeprops={
        'edgecolor': 'black',
        'linewidth': 1 
    }
)
plt.title('Market share of mobile OS')
plt.tight_layout()
plt.show()
```

```python
#data with others expanded 
data = [72, 27, 1]
labels = ['Android', 'iOS', 'Other']
colors = ['#E5CFF7','#713ABE','#5B0888']
#pie chart
plt.pie(
    x = data,
    labels = labels,
    colors = colors,
    shadow = True,
    autopct = '%.1f%%',
    frame = False,
    wedgeprops={
        'edgecolor': 'black',
        'linewidth': 1 
    }
)
plt.title('Market share of mobile OS')
plt.tight_layout()
plt.show()
```

## still data is a bit cluttered , using horizontal bar graph now

```python
data = [72, 27, 0.4, 0.3, 0.2, 0.1]
labels = ['Android', 'iOS', 'KaiOS', 'Windows Phone', 'Samsung', 'Other']
colors = ['#304852','#36595F','#5F818A','#405559','#C7EEFF','#B9DBE6']
plt.barh( labels,width = data, color = colors)
plt.title('Market share of mobile OS')
plt.tight_layout()
plt.show()
```

```python
#data with others expanded 
data = [72, 27, 1]
labels = ['Android', 'iOS', 'Other']
colors = ['#E5CFF7','#713ABE','#5B0888']
plt.barh( labels,width = data, color = colors)
plt.title('Market share of mobile OS')
plt.tight_layout()
plt.show()
```

## conclusion 
- Data with multiple entries like we have labels which are multiple more than three or four, it clutter the pie chart. While in the bar chart, it's clearly visible, so we should prefer bar chat when there are data mode than few.
- and when we used android iOS and others for that, I think we should prefer pie chart since it's more better than the horizontal bar chart
