```python
#imports 
import matplotlib.pyplot as plt 
#data
# Hours studied by each student
hours_studied = [1, 2, 2.5, 3, 3.5, 4, 4.5, 5, 5.5, 6, 6.5, 7, 7.5, 8, 9]

# Corresponding exam scores
exam_scores = [52, 55, 57, 60, 61, 65, 67, 70, 73, 75, 77, 80, 83, 85, 90]

sizes = [score * 2 for score in exam_scores]
colors = ['red' if score < 60 else 'green' for score in exam_scores]

plt.scatter(hours_studied,exam_scores, c=colors, s=sizes , alpha = 0.7)
plt.title('Study Hours vs Exam Score')
plt.xlabel('Study hours')
plt.ylabel('Exam Scores')
plt.grid(True,linestyle='--',color='blue',alpha = 0.3)
plt.tight_layout()
plt.show()
```

## Using a Colormap

```python

#imports 
import matplotlib.pyplot as plt 
import numpy as np
#data
# Hours studied by each student
hours_studied = [1, 2, 2.5, 3, 3.5, 4, 4.5, 5, 5.5, 6, 6.5, 7, 7.5, 8, 9]

# Corresponding exam scores
exam_scores = [52, 55, 57, 60, 61, 65, 67, 70, 73, 75, 77, 80, 83, 85, 90]

sizes = [score * 2 for score in exam_scores]
# colors = ['red' if score < 60 else 'green' for score in exam_scores]
normalized_colors  = np.array(exam_scores)
plt.scatter(hours_studied,exam_scores, c=normalized_colors, s=sizes ,cmap='YlOrRd', alpha = 0.7)
plt.colorbar(label='Score')
plt.title('Study Hours vs Exam Score')
plt.xlabel('Study hours')
plt.ylabel('Exam Scores')
plt.grid(True,linestyle='--',color='blue',alpha = 0.3)
plt.tight_layout()
plt.show()
```

```python
# Assume two groups: Class A and Class B
class_a_hours = [2, 4, 6, 8]
class_a_scores = [45, 55, 65, 85]
class_b_hours = [1, 3, 5, 7, 9]
class_b_scores = [40, 50, 60, 70, 90]
plt.scatter(class_a_hours, class_a_scores, label='Class A', color='blue')
plt.scatter(class_b_hours, class_b_scores, label='Class B', color='orange')
plt.title('Scatter Plot: Class A vs Class B')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.legend()
plt.grid(True)
plt.show()
```
