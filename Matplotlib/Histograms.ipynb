```python
# data of math scores of a class
import matplotlib.pyplot as plt
import numpy as np
#data
math_scores = [0, 96, 67, 68, 70, 72, 74, 75, 77, 78,
               79, 80, 81, 82, 23, 54, 85, 36, 87, 88,
               89, 90, 31, 92, 93, 54, 35, 96, 97, 32]
bins = [0,20,40,60,80,100]
# colors = ['skyblue', 'lightgreen', 'gold', 'lightcoral']
plt.hist(math_scores,bins = bins , alpha = 0.5,color='gold',edgecolor='black')
plt.title('Scores in Mathematics')
plt.ylabel('Number of Students')
plt.xlabel('Marks Scored')
plt.grid(True, color='brown',linestyle='--',alpha=0.6)
plt.axvline(np.mean(math_scores), color='red', linestyle='--', linewidth=2,label='Average Marks')
plt.legend()
plt.show()
```
